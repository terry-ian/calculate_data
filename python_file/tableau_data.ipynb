```python
emptyalldata= pd.DataFrame(columns=['user_id','username','valid_bet','payoff','kind_name','vendor_name','game_name'])
for k in time_total:
    timeall=k  
    game_list_all=[]
    for j in range(list_platform.shape[0]):   #個平台代號
        vendor=list_platform[['id','kind']].iloc[j][0]
        kind=list_platform[['id','kind']].iloc[j][1]
        list_data_all=[]                          
        ttt_total = session_requests.get("https://yb01.88lard.com/api/v1/wagers?vendor="+str(vendor)+"&kind="+str(kind)+"&to_CNY=true&start_at="+timeall+"T00%3A00%3A00-04%3A00&end_at="+timeall+"T23%3A59%3A59-04%3A00&first_result=0&max_results=1&sort=at&sort=id&order=desc&order=desc" , headers=headers )
        ttt_total=int(ttt_total.json()['pagination']['total'])
        for i in range(0, ttt_total, 1000) :   
            ttt = session_requests.get("https://yb01.88lard.com/api/v1/wagers?settled=1&unsettled=0&canceled=1&to_CNY=true&start_at="+timeall+"T00%3A00%3A00-04%3A00&end_at="+timeall+"T23%3A59%3A59-04%3A00&first_result="+str(i)+"&max_results=1000&sort=at&sort=id&order=desc&order=desc&vendor="+str(vendor)+"&kind="+str(kind)+"" , headers=headers )
            output_data=ttt.json()['ret']
            list_data_all=list_data_all+output_data
            print(i)
        game_list_all=game_list_all+list_data_all
        print(vendor)
        session_requests = requests.session()
        response = session_requests.request('PUT',url=LOGIN_URL, data=payload, headers=headers)
    game_data=pd.DataFrame(game_list_all)[['user_id','username','valid_bet','payoff','kind_name','vendor_name','game_name']]
    game_data['Datetime']=timeall
    emptyalldata=emptyalldata.append(game_data)
```

```python
game_data=emptyalldata
```

```python
game_data[['valid_bet','payoff']]=game_data[['valid_bet','payoff']].astype('float')
game_data['payoff']=game_data['payoff']*(-1)
game_data['wager']=0
game_data=game_data.groupby(['Datetime','user_id','username','kind_name','vendor_name','game_name'], as_index=False).agg({
         'valid_bet':sum,    # Sum duration per group
         'payoff': sum,  # get the count of networks
         'wager': 'count'  # get the first date per group
    })
game_data.columns = ['Datetime','user_id','username','kind_name','vendor_name','game_name','valid_bet','payoff','wager']
```

```python
game_data=pd.merge(game_data,VIP_all_user_total,on = 'user_id',how = 'left')
game_data['name']=game_data['name'].fillna('VIP 0')
```

```python
game_data.to_csv('F:/Desktop/tableau_data/usergame_data.csv' ,encoding="utf_8_sig" )
```
